## Load Original 12 datasets 

```python
%%time
airline1 = pd.read_csv("201901_time.csv")
airline2 = pd.read_csv("201902_time.csv")
airline3 = pd.read_csv("201903_time.csv")
airline4 = pd.read_csv("201904_time.csv")
airline5 = pd.read_csv("201905_time.csv")
airline6 = pd.read_csv("201906_time.csv")
airline7 = pd.read_csv("201807_time.csv")
airline8 = pd.read_csv("201808_time.csv")
airline9 = pd.read_csv("201809_time.csv")
airline10 = pd.read_csv("201810_time.csv")
airline11 = pd.read_csv("201811_time.csv")
airline12 = pd.read_csv("201812_time.csv")
# data_dictionary = pd.read_csv("On_time_README.csv")
```

```python
%%time
airlines = pd.concat([airline1
                     ,airline2
                     ,airline3
                     ,airline4
                     ,airline5
                     ,airline6
                     ,airline7
                     ,airline8
                     ,airline9
                     ,airline10
                     ,airline11
                     ,airline12], ignore_index=True)

print(airlines.shape)
```

### Devided into "CANCELL" & "NOT CANCELL"

```python
%%time

def devideCANCELL(filename):
    filename = filename.drop_duplicates()
    
    filename_cancell = filename[(filename['DEP_TIME'].isna() == True) &
                                (filename['ARR_TIME'].isna() == True) & 
                                (filename['CANCELLED'] == 1.0)]
    
    filename_not_cancell = filename [(filename['DEP_TIME'].isna() == False) &
                                     (filename['ARR_TIME'].isna() == False) & 
                                     (filename['CANCELLED'] == 0.0)]
    filename_not_cancell = filename_not_cancell.dropna()
    
    
    filename_cancell.reset_index(drop=True,inplace=True) 
    filename_not_cancell.reset_index(drop=True,inplace=True) 

    print(filename_cancell.shape, filename_not_cancell.shape)
    return filename_cancell, filename_not_cancell

airlines_cancell, airlines_not_cancell = devideCANCELL(airlines)
```

### Float to Int

```python
%%time

def float_To_int(filename, floatColumn):

    length = len(floatColumn)
    for i in range(length):
        column = floatColumn[i]
        filename[column] = filename[column].astype('int')
    return filename


floatColumn = ['DEP_TIME', 'DEP_DELAY', 'DEP_DEL15',
               'ARR_TIME', 'ARR_DELAY', 'ARR_DEL15',
               'CANCELLED', 'DISTANCE']

# new_airline1_not_cancell[new_airline1_not_cancell['ARR_DELAY'].isna()==True]
# new_airline1_not_cancell.head()

airlines_not_cancell = float_To_int(airlines_not_cancell, floatColumn)
print(airlines_not_cancell.shape)
```

## Deal with Hour

```python
%%time

# 把hhmm分离成hh mm两列


def hhmmTo(filename):
    filename['CRS_DEP_HOUR'] = np.floor(filename['CRS_DEP_TIME'].values / 100).astype('int')
    filename['CRS_DEP_MIN'] = filename['CRS_DEP_TIME'].values % 100
    filename['ACT_DEP_HOUR'] = np.floor(filename['DEP_TIME'].values / 100).astype('int')
    filename['ACT_DEP_MIN'] = filename['DEP_TIME'].values % 100
    
    filename['CRS_ARR_HOUR'] = np.floor(filename['CRS_ARR_TIME'].values / 100).astype('int')
    filename['CRS_ARR_MIN'] = filename['CRS_ARR_TIME'].values % 100
    filename['ACT_ARR_HOUR'] = np.floor(filename['ARR_TIME'].values / 100).astype('int')
    filename['ACT_ARR_MIN'] = filename['ARR_TIME'].values % 100
    
    filename = filename.drop(columns=['CRS_DEP_TIME', 'DEP_TIME', 'CRS_ARR_TIME', 'ARR_TIME'])
    
    return filename

airlines_not_cancell = hhmmTo(airlines_not_cancell)
```

# GROUP BY DATE

```python
gbDate = airlines_not_cancell.groupby(['DATE'])

gbDate_mean = gbDate.mean()[['DEP_DELAY','ARR_DELAY']]
gbDate_mean.rename(columns={'DEP_DELAY':'DEP_MEAN', 'ARR_DELAY':'ARR_MEAN'}, inplace=True)

gbDate_min = gbDate.min()[['DEP_DELAY','ARR_DELAY']]
gbDate_min.rename(columns={'DEP_DELAY':'DEP_MIN', 'ARR_DELAY':'ARR_MIN'}, inplace=True)

gbDate_max = gbDate.max()[['DEP_DELAY','ARR_DELAY']]
gbDate_max.rename(columns={'DEP_DELAY':'DEP_MAX', 'ARR_DELAY':'ARR_MAX'}, inplace=True)

gbDate_result = pd.concat([gbDate_mean, gbDate_min, gbDate_max], axis=1)
gbDate_result.to_csv('gbDate.csv')

gbDate_result
```
